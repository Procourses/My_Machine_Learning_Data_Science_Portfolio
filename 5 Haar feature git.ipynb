```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2               

image = cv2.imread('gov.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  
plt.figure(figsize=(20,10))
plt.imshow(gray, cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray, 1.06, 1)

print ('found ' + str(len(faces)) + ' faces in image')
print ("Their coordinates and lengths/widths are:")
print ('=============================')
print (faces)
```

```python
img_with_detections = np.copy(image)   
for (x,y,w,h) in faces:
    cv2.rectangle(img_with_detections,(x,y),(x+w,y+h),(255,0,0),2)  
plt.figure(figsize=(20,10))
plt.imshow(img_with_detections)
```
